## Importing important libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import sklearn.preprocessing as preprocessing
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
pd.get_dummies(df, prefix='encoded')
```

```python
target_var = df['Survived'] 
cols = list(df.columns)
cols.remove("Survived")
features = df[cols]
```

```python
model = linear_model.LinearRegression()
```

```python
# Dividing sample and test data 
from sklearn.model_selection import train_test_split

features_train, features_test, target_train, target_test = train_test_split(features, target_var, test_size=0.10, random_state=1)
```

```python
def summarise(df):
    for column in df.columns:
        print(column)
        if df.dtypes[column] == np.object:
            print("{}\n".format(df[column].value_counts()))
        else:
            print("{}\n".format(df[column].describe()))
        
        
summarise(df)
```

```python
# Encode the data using LabelEncoder

def encodeData(df):
    result = df.copy()
    encoders = {}
    for column in df.columns:
        if df.dtypes[column] == np.object:
            encoders[column] = preprocessing.LabelEncoder()
            result[column] = encoders[column].fit_transform(result[column])
    return result, encoders

encoded_data, _ = encodeData(df)

print(encoded_data.head())
```

```python
def roundingup(df):
    for col in df.columns:
        print(df[col].dtype)
        if df[col].dtype == np.float64:
            print("y")
    return None
            
roundingup(df)
```

```python
# Teaching the model

model.fit(features_train, target_train)
```
