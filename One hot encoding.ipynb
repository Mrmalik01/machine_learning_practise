```python
import pandas as pd
from IPython.display import display

data = pd.DataFrame({
    "City" : ['London', 'York', 'Gloucestershire', 'Manchester'],
    "Value" : [9, 5, 3, 8]
})

display(data)

# One hot encoding using pd.get_dummies method


encoded = pd.get_dummies(data,prefix=['City'])

display(encoded)
```
